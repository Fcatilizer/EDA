```python
#changing null values to mean in df
mean_df = df.fillna(df.mean())
```

```python
mean_df.isnull().sum()
```

```python
median_df = df.fillna(df.median())
```

```python
median_df.isnull().sum()
```

```python
mean_df.describe()
```

```python
#adding new mean_age column to df
df['mean_age'] = mean_df['Age'] 
```

```python
df['median_age'] = median_df['Age']
```
